```python
import sqlite3 as sql
import urllib
import requests
from urllib import parse
from bs4 import BeautifulSoup
```

```python
try:
    con = sql.connect('air.db')
    cur = con.cursor()
    cur.execute("DROP TABLE IF EXISTS Polution")
    cur.execute("CREATE TABLE Polution (Datatime TEXT primary key, so2Value TEXT, coValue TEXT, o3Value TEXT, no2Value TEXT, pm10Value TEXT, pm10Value24 TEXT, pm25Value TEXT, pm25Value24 TEXT, khaiValue TEXT, khaiGrade TEXT, so2Grade TEXT, coGrade TEXT, o3Grade TEXT, no2Grade TEXT, pm10Grade TEXT, pm25Grade TEXT, pm10Grade1h TEXT, pm25Grade1h TEXT)")
except Exception as e:
    print("Error: ", e)
```
