```python
from pymongo import MongoClient
```

```python
client = MongoClient('localhost', 27017)
```

```python
total = client.get_database('total_db')
```

```python
diaper = total.get_collection('Diaper')
```

```python
diaperResult = diaper.find({})
```

```python
diaperList = list(diaperResult)
```

```python
len(diaperList)
```

```python
for i in range(len(diaperList)):
    if type(diaperList[i]['price']) == str:
        diaperList[i]['price'] = int(diaperList[i]['price'].replace(',', '').replace('\n', '').replace('Ïõê', '').replace('~', ''))
```

```python
count = 0

for item in diaperList:
    print(type(item['price']))
```

```python
for item in diaperList:
    if type(item['replyCount']) == str:
        print(item['replyCount'])
```

```python
for item in diaperList:
    if item['replyCount'] == 'No reply':
        item['replyCount'] = 0
```

```python
for item in diaperList:
    if type(item['replyCount']) == str:
        item['replyCount'] = int(item['replyCount'].replace('Í∞úÏùò ÌõÑÍ∏∞', '').replace(',', ''))
```

```python
for item in diaperList:
    if item['rating'] == None:
        item['rating'] = 0
```

```python
for item in diaperList:
    if item['rating'] == 'No rating':
        item['rating'] = 0
```

```python
for item in diaperList:
    item['rating'] = float(item['rating'])
```

```python
for item in diaperList:
    print(item['rating'])
```

```python
for item in diaperList:
    print(type(item['replyCount']))
```

```python
diaper_collection = total.get_collection('Diaper')
```

```python
diaper_collection.insert_many([
    diaperList[i] for i in range(len(diaperList))
])
```
